```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', 100)
pd.set_option('display.float_format', '{:.2f}'.format)

print("Libraries loaded ✅")
```

## 1. Load Dataset

```
# If running on Colab, upload application_train.csv from Kaggle
# from google.colab import files
# uploaded = files.upload()

df = pd.read_csv('application_train.csv')
print(f"Shape: {df.shape}")
df.head()
```

## 2. Initial Assessment

```
# Basic info
print("=== Dataset Info ===")
print(f"Rows: {df.shape[0]:,}")
print(f"Columns: {df.shape[1]}")
print(f"\nTarget distribution:")
print(df['TARGET'].value_counts())
print(f"\nDefault rate: {df['TARGET'].mean()*100:.2f}%")
```

```
# Missing values analysis
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
}).query('`Missing Count` > 0').sort_values('Missing %', ascending=False)

print(f"Columns with missing values: {len(missing_df)}")
missing_df.head(20)
```

```
# Data types overview
print("Numeric columns:", df.select_dtypes(include='number').shape[1])
print("Object columns:", df.select_dtypes(include='object').shape[1])
```

## 3. Drop High-Missing Columns (>60% missing)

```
# Drop columns with more than 60% missing
threshold = 0.6
cols_to_drop = missing_df[missing_df['Missing %'] > 60].index.tolist()
print(f"Dropping {len(cols_to_drop)} columns with >60% missing values")
print(cols_to_drop[:10], "...")

df.drop(columns=cols_to_drop, inplace=True)
print(f"\nNew shape: {df.shape}")
```

## 4. Handle Missing Values

```
# Numeric columns: fill with median
numeric_cols = df.select_dtypes(include='number').columns.tolist()
numeric_cols = [c for c in numeric_cols if c != 'TARGET']

for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

# Categorical columns: fill with mode
cat_cols = df.select_dtypes(include='object').columns.tolist()
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

print(f"Remaining nulls: {df.isnull().sum().sum()}")
```

## 5. Fix Data Types & Negative Values

```
# DAYS columns are negative in this dataset (days before application)
# Convert to positive years
days_cols = [c for c in df.columns if c.startswith('DAYS_')]
print("DAYS columns:", days_cols)

for col in days_cols:
    df[col] = df[col].abs()

# Create AGE in years from DAYS_BIRTH
df['AGE'] = (df['DAYS_BIRTH'] / 365).astype(int)
print(f"\nAge range: {df['AGE'].min()} - {df['AGE'].max()} years")
```

## 6. Feature Engineering

```
# AGE_GROUP
def age_group(age):
    if age < 25: return '<25'
    elif age < 35: return '25-35'
    elif age < 45: return '35-45'
    elif age < 55: return '45-55'
    else: return '55+'

df['AGE_GROUP'] = df['AGE'].apply(age_group)
print("AGE_GROUP distribution:")
print(df['AGE_GROUP'].value_counts())
```

```
# INCOME_BRACKET
def income_bracket(income):
    if income < 90000: return '<90K'
    elif income < 180000: return '90K-180K'
    elif income < 270000: return '180K-270K'
    elif income < 400000: return '270K-400K'
    else: return '400K+'

df['INCOME_BRACKET'] = df['AMT_INCOME_TOTAL'].apply(income_bracket)
print("INCOME_BRACKET distribution:")
print(df['INCOME_BRACKET'].value_counts())
```

```
# ANNUITY_INCOME_RATIO
df['ANNUITY_INCOME_RATIO'] = df['AMT_ANNUITY'] / df['AMT_INCOME_TOTAL']

# CREDIT_INCOME_RATIO
df['CREDIT_INCOME_RATIO'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']

print("New features created ✅")
print(df[['AGE_GROUP','INCOME_BRACKET','ANNUITY_INCOME_RATIO','CREDIT_INCOME_RATIO']].describe())
```

## 7. Remove Outliers

```
# Cap extreme outliers using IQR on key numeric columns
def cap_outliers(df, col):
    Q1 = df[col].quantile(0.01)
    Q3 = df[col].quantile(0.99)
    df[col] = df[col].clip(lower=Q1, upper=Q3)
    return df

for col in ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY']:
    df = cap_outliers(df, col)

print("Outliers capped ✅")
```

## 8. Final Check & Export

```
print("=== Final Dataset Summary ===")
print(f"Shape: {df.shape}")
print(f"Nulls remaining: {df.isnull().sum().sum()}")
print(f"Target distribution:\n{df['TARGET'].value_counts()}")
print(f"Default rate: {df['TARGET'].mean()*100:.2f}%")
df.head()
```

```
# Export cleaned dataset
df.to_csv('application_clean.csv', index=False)
print("Exported: application_clean.csv ✅")

# Preview key engineered columns
print("\nNew columns added:")
print([c for c in df.columns if c not in pd.read_csv('application_train.csv', nrows=1).columns])
```
